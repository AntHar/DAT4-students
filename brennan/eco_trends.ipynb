# Modeling economic trends with Quandl data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import Quandl 
```

### Locate daily indices representative of market forces:

```python
#see Quandl.com
gold = "FRED/GOLDPMGBD228NLBM" #since 1968
oil = "DOE/RWTC" #since 1986
rates = "FRED/DFF" #since 1954
stocks = "YAHOO/INDEX_GSPC" #since 1950
dollar = "FRED/DTWEXM" #since 1973
forces = [stocks, rates, dollar, gold, oil]
```

### Load dataframe and check row length:

```python

df = Quandl.get(forces, authtoken=token, trim_start='1987-01-01')
#df.isnull().sum()
len(df)
```

### Trim fluff and clean up column names:

```python
col = list(df.columns.values)
for i in col[:4]:
    df.drop(i, axis=1, inplace=True)
    
df.rename(columns={'YAHOO.INDEX_GSPC - Volume' : 'vol - Volume'}, inplace=True)

col = list(df.columns.values)
for i in col:
    tag = i.split(' ')[0][-3:].lower()
    df.rename(columns={i : tag}, inplace=True)
    
df.head(3)
```

### Normalize columns:

```python
dfn = df.copy()
dfn = (dfn - dfn.mean()) / (dfn.max() - dfn.min())
```

### Plot normalized data:

```python
col = list(dfn.columns.values)
plt.clf()
fig = plt.gcf()
fig.set_size_inches(14, 6)
for i in col:
    plt.plot(dfn.index, dfn[i])
plt.legend(col, loc='upper center', bbox_to_anchor=(0.3, 1),
          ncol=6, fancybox=True, shadow=True)
plt.title("Normalized indicators")
```

### Calculating the convergence/divergence of simple moving averages:

```python
def find_smd(d, lng, sht):
    smd = pd.stats.moments.rolling_mean(d, lng) - pd.stats.moments.rolling_mean(d, sht)
    return smd
```

```python
col = list(dfn.columns.values)
dfnSmd = dfn.copy()
for i in col:
    dfta = dfnSmd[i].copy()
    dfta = dfta.dropna()
    dfta = find_smd(dfta, 200, 50)
    dfnSmd[i] = dfta

col = list(dfnSmd.columns.values)

plt.clf()
fig = plt.gcf()
fig.set_size_inches(14, 6)
for i in col:
    plt.plot(dfnSmd.index, dfnSmd[i])
plt.legend(col, loc='upper center', bbox_to_anchor=(0.3, 1),
          ncol=6, fancybox=True, shadow=True)
plt.title("Simple moving average differential: 200 vs 50 day")
```

### Calculating the exponentially weighted MACD (moving avg. convergence/divergence):

```python
def find_emd(d, lng, sht):
    emd = pd.stats.moments.ewma(d , span=lng, freq="D") - pd.stats.moments.ewma(d , span=sht, freq="D")
    return emd
```

```python
col = list(dfn.columns.values)
dfnEmd = dfn.copy()
for i in col:
    dfta = dfnEmd[i].copy()
    #dfta = dfta.dropna()
    dfta = find_emd(dfta, 26, 12)
    dfnEmd[i] = dfta

col = list(dfnEmd.columns.values)
plt.clf()
fig = plt.gcf()
fig.set_size_inches(14, 6)
for i in col:
    plt.plot(dfnEmd.index, dfnEmd[i])
plt.legend(col, loc='upper center', bbox_to_anchor=(0.3, 1),
          ncol=6, fancybox=True, shadow=True)
plt.title("MACD differential: 26 vs. 12 day")
```

```python
#Smoother for better visuals

def make_smooth(d, period):
    smooth = pd.stats.moments.rolling_mean(d, period)
    return smooth
```

```python
col = list(dfnEmd.columns.values)
dfnEmdSmooth = dfnEmd.copy()
for i in col:
    dfta = dfnEmdSmooth[i].copy()
    dfta = dfta.dropna()
    dfta = make_smooth(dfta, 600)
    dfnEmdSmooth[i] = dfta
dfnEmdSmooth.head(2)

col = list(dfnEmdSmooth.columns.values)
#col = col[-5:]
plt.clf()
fig = plt.gcf()
fig.set_size_inches(14, 6)
for i in col:
    plt.plot(dfnEmdSmooth.index, dfnEmdSmooth[i])
plt.legend(col, loc='upper center', bbox_to_anchor=(0.3, 1),
          ncol=6, fancybox=True, shadow=True)

plt.title("MACD differential: 26 vs. 12 day - smoothed")
```

### Calculating bollinger values: 

```python
def find_bolval(d, span): #redo with ewma?
    ma = pd.stats.moments.rolling_mean(d, span)  #d['MA20']
    ab = abs(d - ma)  #d['ABS']
    sd = pd.stats.moments.rolling_std(ab, span)  #d['STDDEV']
    #d['UPPER'] = d['MA20'] + 2*d['STDDEV']  
    #d['LOWER'] = d['MA20'] - 2*d['STDDEV']
    #d['bot_ref'] = d - d['LOWER']
    d = (d - ma) / sd
    return d
```

```python
col = list(dfn.columns.values)
dfnBol = dfn.copy()
for i in col:
    dfta = dfnBol[i].copy()
    dfta = dfta.dropna()
    dfta = find_bolval(dfta, 20)
    dfnBol[i] = dfta
dfnBol.head(5)

col = list(dfnBol.columns.values)

plt.clf()
fig = plt.gcf()
fig.set_size_inches(14, 6)
for i in col:
    plt.plot(dfnBol.index, dfnBol[i])
plt.legend(col, loc='upper center', bbox_to_anchor=(0.3, 1),
          ncol=6, fancybox=True, shadow=True)
plt.title("Bollinger values: 20 day")
```

```python
#Now smoother

col = list(dfnBol.columns.values)
dfnBolSmooth = dfnBol.copy()
for i in col:
    dfta = dfnBolSmooth[i].copy()
    dfta = dfta.dropna()
    dfta = make_smooth(dfta, 600)
    dfnBolSmooth[i] = dfta
#dfnEmdSmooth.head(2)

col = list(dfnBolSmooth.columns.values)
#col = col[-5:]
plt.clf()
fig = plt.gcf()
fig.set_size_inches(14, 6)
for i in col:
    plt.plot(dfnBolSmooth.index, dfnBolSmooth[i])
plt.legend(col, loc='upper center', bbox_to_anchor=(0.3, 1),
          ncol=6, fancybox=True, shadow=True)
plt.title("Bollinger values: 20 day - smoothed")
```
